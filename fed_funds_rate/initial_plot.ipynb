```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read data
df = pd.read_csv("data.csv", index_col = 0)
df.head()
```

```python
df.plot(figsize = (24,10))
plt.title("Initial Plot")
plt.grid()
```

```python
df = df.drop(columns = ["IRLTLT01EZM156N"])
```

```python
df.plot(figsize = (24,10))
plt.title("Initial Plot")
plt.grid()
```

```python
#now for the plot
plt.plot(df[df.columns[0]], df[df.columns[1]], "o")
plt.show()
plt.plot(df[df.columns[0]], df[df.columns[2]], "o")
plt.show()
plt.plot(df[df.columns[0]], df[df.columns[3]], "o")
plt.show()
```

```python
#let's add in axes
plt.plot(df[df.columns[0]], df[df.columns[1]], "o")
plt.axhline(y = 0, color = "r", linestyle = "-")
plt.axvline(x = 0, color = "r", linestyle = "-")
plt.show()
plt.plot(df[df.columns[0]], df[df.columns[2]], "o")
plt.axhline(y = 0, color = "r", linestyle = "-")
plt.axvline(x = 0, color = "r", linestyle = "-")
plt.show()
plt.plot(df[df.columns[0]], df[df.columns[3]], "o")
plt.axhline(y = 0, color = "r", linestyle = "-")
plt.axvline(x = 0, color = "r", linestyle = "-")
plt.show()
```

```python
#let's loop through so that the code is nicer

#we want to loop through the last 3 columns which is one less than the number we have already
for i in range(len(df.columns) - 1):
    
    #we are starting at the i+1 spot
    plt.plot(df[df.columns[0]], df[df.columns[i+1]], "o")
    plt.axhline(y = 0, color = "r", linestyle = "-")
    plt.axvline(x = 0, color = "r", linestyle = "-")
    plt.title(df.columns[i+1])
    plt.show()
```

```python
#let's rename the columns so that they are nicer
df.columns = ["FEDFUNDS", "UK", "US", "JPY"]
df.head()
```

```python
#we want to loop through the last 3 columns which is one less than the number we have already
for i in range(len(df.columns) - 1):
    
    #we are starting at the i+1 spot
    plt.plot(df[df.columns[0]], df[df.columns[i+1]], "o")
    plt.axhline(y = 0, color = "r", linestyle = "-")
    plt.axvline(x = 0, color = "r", linestyle = "-")
    plt.title(df.columns[i+1])
    plt.show()
```

```python
#let's put it all one one plot
fig, axes = plt.subplots(1, 3, figsize = (30, 10))

for i in range(len(df.columns) - 1):
    
    axes[i].plot(df[df.columns[0]], df[df.columns[i+1]], "o")
    axes[i].axhline(y = 0, color = "r", linestyle = "-")
    axes[i].axvline(x = 0, color = "r", linestyle = "-")
    axes[i].set_title(df.columns[i+1])
    axes[i].set_xlabel("Fed Funds Target Rate Change")
    axes[i].set_ylabel("Government Bond Return Percent Change (10YR)")
    axes[i].grid()
    
plt.suptitle("Monthly Change Fed Funds Rate vs Monthly Change in Government Bonds (10Yr) from {} to {}".format(df.index[0], df.index[len(df) - 1]))
plt.tight_layout()
```
